# **HOFFMAN**

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import sys
from google.colab import files

def lz77_compress(data):
    compressed_data = []
    i = 0
    while i < len(data):
        match_len = 0
        match_dist = 0
        # Find the longest match
        for j in range(i + 1, len(data)):
            match = data[i:j]
            # Ensure substring match in data
            if j + len(match) <= len(data) and np.array_equal(data[j:j + len(match)], match):
                match_len = len(match)
                match_dist = i
            else:
                break

        if match_len > 0:
            compressed_data.append((match_dist, match_len, data[i + match_len]))
            i += match_len + 1
        else:
            compressed_data.append((0, 0, data[i]))
            i += 1

    return compressed_data


def lz77_decompress(compressed_data):
    decompressed_data = []
    for dist, length, char in compressed_data:
        if length > 0:
            start = len(decompressed_data) - dist
            decompressed_data.extend(decompressed_data[start:start + length])
        decompressed_data.append(char)

    return decompressed_data


def compress_image(image_path):
    img = Image.open(image_path)
    img = img.convert('L')  # Convert to grayscale

    pixels = np.array(img)
    flat_pixels = pixels.flatten()

    # Compress the pixel data using LZ77
    compressed_data = lz77_compress(flat_pixels)
    return compressed_data, pixels.shape, pixels


def decompress_image(compressed_data, original_shape):
    decompressed_data = lz77_decompress(compressed_data)
    decompressed_pixels = np.array(decompressed_data).reshape(original_shape)

    return decompressed_pixels


def save_image(image_data, output_path):
    image = Image.fromarray(image_data.astype(np.uint8))
    image.save(output_path)
    print(f"Image saved at {output_path}")


def display_image(image_data, title="Image"):
    plt.imshow(image_data, cmap='gray')
    plt.axis('off')  # Hide axes
    plt.title(title)
    plt.show()


def main():
    # Upload file directly to Colab
    print("Please upload the image file:")
    uploaded = files.upload()

    # Get the uploaded file path
    image_path = list(uploaded.keys())[0]
    print(f"File uploaded: {image_path}")

    # Compress the image
    compressed_data, original_shape, original_pixels = compress_image(image_path)
    print("Image compressed using LZ77!")

    # Calculate the file sizes
    original_size = sys.getsizeof(original_pixels.flatten())
    compressed_size = sum(sys.getsizeof(item) for item in compressed_data)

    # Calculate the compression ratio
    compression_ratio = original_size / compressed_size if compressed_size > 0 else float('inf')
    print(f"Original size: {original_size / 1024:.2f} KB")
    print(f"Compressed size: {compressed_size / 1024:.2f} KB")
    print(f"Compression ratio: {compression_ratio:.2f}")

    # Display the original image before decompression
    display_image(original_pixels, title="Original Image")

    # Decompress the image
    decompressed_pixels = decompress_image(compressed_data, original_shape)

    # Display the decompressed image
    display_image(decompressed_pixels, title="Decompressed Image")

    # Save the decompressed image
    output_path = "decompressed_image.png"
    save_image(decompressed_pixels, output_path)


if __name__ == "__main__":
    main()
```
